```python
from bokeh.plotting import figure, output_file, show

df = pd.read_csv('sportspersons.csv')
height = df.iloc[:,3]
weight = df.iloc[:,4]
output_file('sportsmen_h_w.html')
x = height
y = weight
p = figure()
p.circle(x, y, size=5, color='red', legend='circle')
p.xaxis.axis_label = 'Height'
p.yaxis.axis_label = 'Weight'
show(p)
```

```python
import pandas as pd
import bokeh.plotting as bpl
import bokeh.models as bmo
from bokeh.palettes import d3
bpl.output_notebook()


df = pd.DataFrame(
    {
       
        "cat": ['a', 'b', 'a', 'c' ],
        "kpi1": [1,2,3,4],
        "kpi2": [4,3,2,1]
    }
)

print(df['cat'].unique())
source = bpl.ColumnDataSource(df)

# use whatever palette you want...
print(d3)
palette = d3['Category10'][len(df['cat'].unique())]
color_map = bmo.CategoricalColorMapper(factors=df['cat'].unique(),
                                   palette=palette)

# create figure and plot
p = bpl.figure()
p.scatter(x='kpi1', y='kpi2',
          color={'field': 'cat', 'transform': color_map},
          legend='cat', source=source)
bpl.show(p)
```
